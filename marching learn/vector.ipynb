```python
import numpy as np
```

```python
# 定义向量
x = np.array([2, 6, 8])
print(x)
print(x.shape)
```

```python
# 向量转置
print(x.T)
print(x.shape)
```

```python
# 相加
y = np.array([1, 3, 7])
print(x + y)
```

```python
# 向量与标量相乘
print(x * 3)
```

```python
# 向量与向量相乘 (对应位置元素相乘)
print(x * y)
```

```python
# 向量点乘
# np.dot(x, y)
# x @ y
print(x.dot(y))
```

```python
# 计算范数
l0_norm = np.linalg.norm(x, ord=0)
print(l0_norm)
l1_norm = np.linalg.norm(x, ord=1)
print(l1_norm)
l2_norm = np.linalg.norm(x, ord=2)
print(l2_norm)
```
