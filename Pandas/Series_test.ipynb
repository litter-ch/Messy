```python
np.random.seed(1)
hours_sales = pd.Series(np.random.randint(0, 100, 24),
                        index=pd.date_range('2025-01-01', periods=24, freq='h'))
print(hours_sales)
```

```python
# 按天重采样计算每日总销售额
day_sales = hours_sales.resample('D').sum()
print(day_sales)
# 计算每天营业时间(8:00-22:00)和非营业时间的销售额比例
# hours_sales.between_time('8:00','22:00').sum()
business_hours_sales = hours_sales[(hours_sales.index.hour >= 8) & (hours_sales.index.hour <= 22)].sum()
print(business_hours_sales / (day_sales - business_hours_sales))
# 找出销售最高的三小时
print(hours_sales.nlargest(3))
# 销售最低的两小时
print(hours_sales.nsmallest(2))
```
