```python
import pandas as pd
# 数据分箱
df1 = pd.read_csv('data/nba.csv')
df = df1[5:15][['Name', 'Salary']]
df
```

```python
# 等距划分
pd.cut(df['Salary'], bins=3)  # bins=n, 分成n段区间
#1160130.96 ~ 6585480.0 ~ 12000000.0
pd.cut(df['Salary'], bins=3).value_counts()
pd.cut(df['Salary'], bins=[0, 4000000, 8000000, 12000000])  # bin=list, 分成n段区间
df['收入范围'] = pd.cut(df['Salary'], bins=[0, 4000000, 8000000, 12000000],
                        labels=['低', '中', '高'])
# 数值=>分箱=>统计
# df['gender'] = df['gender'].astype('category')  # 分类类型  字符串=>类别=>统计
# df['gender'].value_counts()
df
```

```python
# 等频划分
pd.qcut(df['Salary'],3).value_counts()
```
