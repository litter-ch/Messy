```python
import pandas as pd

from html01 import index
```

```python
# series创建
s = pd.Series([1, 2, 3, 4, 5])
# 自定义索引
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
# 定义name
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'], name='星期')
print(s)
```

```python
# 通过字典创建
s = pd.Series({'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}, name='星期')
print(s, '\n')
s1 = pd.Series(s, index=['a', 'c'])
print(s1)
```

```python
print('values取到的:', s.values)  # Series的值
print('name取到的:', s.name)  # Series的名字
print(s.loc['a':'b'])  # location 自定义的显式索引 loc可以切片获取范围的元素
print(s.iloc[0:2])  #  系统定义从0开始的隐式索引
print('at取到的:', s.at['c'])  # 显式索引  at只能获取单个特定元素, 但是速度比loc快
print('iat取到的', s.iat[3])  # 隐式索引
```

```python
# 访问数据
s['f'] = 6  # 加一个数据
print(s['f'])
print(s[s < 3])
print(s.head(3))  # 从头部取信息   默认5行
print(s.tail(2))  # 从尾部取信息   默认5行
# print(s.describe()) # 常见统计信息
```

```python
import numpy as np

s = pd.Series([1, 2, np.nan, None, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e', 'f', 'g'], name='data')
print(s)
```

```python
s.describe()  # 查看所有的描述信息
```

```python
# 获取索引
print(s.keys())  # 方法
print(s.index)  # 属性
# s.isna()    # 是否为缺失值
```

```python
s.isin([4, 5, 6])  # 检查每个元素是否在参数集合中
```

```python
print(s.sort_values())  # 值排序    默认ascending=True 从小到大
# print(s.sort_index()) # 索引排序
# print(s.sort_index(ascending=False))
print(s.quantile(0.25))  # 分位数
# print(s.mode()) # 众数
# print(s.value_counts()) # 每个元素出现的次数
```

```python
print(s.drop_duplicates())  # 去重 返回的Series类型
print(s.unique())  # 去重 返回的list列表类型
print(s.nunique())  # 去重之后的元素个数
```
