```python
# DaraFrame的创建
import numpy as np
import pandas as pd
# display()函数, 强制显示表格形式
# 通过Series创建
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([6, 7, 8, 9, 10])
df1 = pd.DataFrame({'第一列':s1,'第二列':s2})
# print(df1)
# 通过字典创建
df = pd.DataFrame(
    {
        'name':['Tom','Jack','Alice','Lucy'],
        'age':[20,22,21,24],
        'gender':['male','male','female','female'],
        'score':[71,62.5,80,86]
    },index=[1,2,3,4],columns=['name','age','gender','score']
)
df
```

```python
# 获取某行元素
# print(df.loc[4]) # 显式索引
print(df.iloc[3])  # 隐式索引
# 获取单个元素
print(df.at[2,'gender'])
print(df.iat[2,1])
print(df.iloc[2,2])
```

```python
# 获取某列元素
# print(df.loc[:,'name'])
print(df.iloc[:,0])
# print(df['gender'])
print(df.gender)
```

```python
# 获取多列数据
print(df[['name','gender']])
print(df.head(2))
print(df.tail(2))
```

```python
# 布尔索引筛选数据
print(df[df.score > 70])
df[(df.score > 60)&(df.age>20)]
```

```python
# 随机抽样
df.sample(2)
```

```python
print(df.head(2))   # 默认查看前5行
print(df.tail(2))   # 默认查看后5行
```

```python
print(df.isin(['Jack',20])) # 查看元素是否包含参数集合里的
print(df.isna())   # 查看元素是否为缺失值
```

```python
print(df.score.sum())   # 总分
print(df.score.mean())  # 平均分
print(df.score.max())   # 最大
print(df.score.median())# 中位数
print(df.score.std())   # 标准差
print(df.score.var())   # 方差
print(df.score.quantile(0.25))  # 分位数
# print(df.describe()) # 详细信息
# df.cummax()   # 累积大
```

```python
# display(df.replace('male','男'))  # 替换
print(df.sort_values(by=['gender','score'], ascending=[False,True])) # 值排序
df.nlargest(2,columns=['score'])    # 取两个最大的score
```
