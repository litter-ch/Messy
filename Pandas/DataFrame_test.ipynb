```python
import pandas as pd
data = {
    '姓名':['张三','李四','王五','周六','李七'],
    '数学':[85,92,78,88,95],
    '英语':[90,88,85,92,80],
    '物理':[75,80,88,85,90]
}
scores = pd.DataFrame(data)
scores
```

```python
scores['总分']=scores[['数学','英语','物理']].sum(axis=1)
# scores['平均分']=scores[['数学','英语','物理','总分']].mean(axis=0)
scores
```
