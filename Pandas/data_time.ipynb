```python
# 时间数据处理
import pandas as pd
d = pd.Timestamp('2019-6-30 14:19:32')
print(d)
print(type(d))
print('年:', d.year)
print('月:', d.month)
print('日:', d.day)
print('时:', d.hour)
print('分', d.minute)
print('秒:', d.second)
print('季度', d.quarter)
print('是否月底:', d.is_month_end)
```

```python
# 方法
print('星期几:', d.day_name())
print('转换为天:', d.to_period('D'))
print('转换为周:', d.to_period('W'))  # 用于数据归类在哪一周
print('转换为月度:', d.to_period('M'))  # 用于数据归类在哪一月
print('转换为年度:', d.to_period('Y'))
print('转换为季度:', d.to_period('Q'))
```

```python
# 字符串转换为日期类型
a = pd.to_datetime('2015-02-28 10:22')
print(a)
print(type(a))
print(a.day_name())
# DataFrame 日期转换
df = pd.DataFrame({
    'sales': [100, 200, 300],
    'data': ['20180506', '20190403', '20200106']
})
df['datetime'] = pd.to_datetime(df['data'])
df['weekday'] = df['datetime'].dt.day_name()
df
```

```python
# 日期数据作为索引
# df.set_index('datetime',inplace=True)
# df.reset_index(inplace=True)
print(df.loc['2018-01':'2019-12'])
```

```python
# 时间间隔
d1 = pd.Timestamp('2019-4-15')
d2 = pd.Timestamp('2025-11-17')
d3 = d2 - d1
print(d3)
print(type(d1))
print(type(d3))
```

```python
# 时间序列
days1 = pd.date_range('2018-09-15', '2019-1-27', freq='W')   # 以周为频率
print(days1)
days2 = pd.date_range('2018-09-15',periods=5, freq='YE')    # periods 生成的时间点数
print(days2)
```

```python
# 日期偏移
date = pd.to_datetime('2008-5-3')
new_date = date + pd.Timedelta(days=50)
print(new_date)
```
