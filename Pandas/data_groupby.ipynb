```python
# 分组聚合
# df.groupby('分组的字段')['聚合的字段'].聚合函数()
import pandas as pd
df = pd.read_csv('data/nba.csv')
df = df.dropna(subset=['Team']) # 清除Team 列的缺失值那一行
print(df.groupby('Team').groups)# 查看分组
df.groupby('Team').get_group('Detroit Pistons')# 查看分组具体的某组数据
```

```python
df2 = df.groupby('Team')[['Salary']].mean().round(2) #分组后平均值聚合保留两位小数
df2 = df2.reset_index()   # 重置索引
df2.sort_values('Salary',ascending=False)   # 值排序  ascending=False从高到低
```
