```python
# 折线图
import matplotlib.pyplot as plt
# plt.rcParams['font.family'] = ['SimHei'] # 字体
plt.rcParams['font.sans-serif'] = ['SimHei']
# 创建图表, 设置大小
plt.figure(figsize=(7,3))

# 绘图的数据
month = ['1月','2月','3月','4月']
sales = [100,150,80,130]
# 绘制折线图
plt.plot(month,sales,label='产品A',color='orange',linewidth=2,marker='o') # marker标记

# 添加标题
plt.title('季度销售趋势',color='green',fontsize=10)
# 添加坐标轴的标签
plt.xlabel('月份',color='green',fontsize=10)
plt.ylabel('销售额(万元)',color='green',fontsize=10)
# 添加图例
plt.legend(loc='upper left')
# 添加网格线
# plt.grid(True)  # x,y 轴都显示
plt.grid(axis='y',alpha=0.1,color='green',linestyle='--')# alpha透明度 #linestyle='--'虚线
# 设置刻度字体大小
plt.xticks(rotation=0,fontsize=9)# rotation旋转角度
plt.yticks(rotation=0,fontsize=9)
# 设置y轴的范围
plt.ylim(0,200)
# 在数据点是显示数据
for x,y in zip(month,sales):
    plt.text(x,y+5,str(y),ha='center',va='bottom',fontsize=10)
# # 自动优化排版
# plt.tight_layout()

# 显示图表
plt.show()
```

```python
# 散点图
import random
plt.rcParams['font.sans-serif'] = ['SimHei']
# 创建图表, 设置大小
plt.figure(figsize=(10,6))

# 绘图的数据
x = []
y = []
for i in range(1000):
    tmp = random.uniform(0,10)
    x.append(tmp)
    tmp2 = tmp*2 + random.gauss(0,2)
    y.append(tmp2)
# 绘制散点图
plt.scatter(x,y,color='blue',alpha=0.5,label='数据',s=20)    # s 圆点大小

# 添加标题
plt.title('x与y的关系',color='green',fontsize=15)
# 添加坐标轴的标签
plt.xlabel('x变量',color='green',fontsize=10)
plt.ylabel('y变量',color='green',fontsize=10)
# 添加图例
plt.legend(loc='upper left')
# 添加网格线
plt.grid(True,alpha=0.1,color='green',linestyle='--')# alpha透明度 #linestyle='--'虚线
# 设置刻度字体大小
plt.xticks(rotation=0,fontsize=9)# rotation旋转角度
plt.yticks(rotation=0,fontsize=9)
# 设置y轴的范围
plt.ylim(0,30)
plt.plot([0,10],[0,20],color='red',linestyle='-')   # 回归线
# 自动优化排版
plt.tight_layout()

# 显示图表
plt.show()
```

```python
# 柱状图
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
# 创建图表, 设置大小
plt.figure(figsize=(7,4))

# 绘图的数据
subjects = ['语文','数学','英语','物理']
scores = [85,92,78,88]

# 绘制柱状图
plt.bar(subjects,scores,label='小会',color='green',width=0.5)

# 添加标题
plt.title('成绩分布',color='green',fontsize=10)
# 添加坐标轴的标签
plt.xlabel('科目',color='green',fontsize=10)
plt.ylabel('分数',color='green',fontsize=10)
# 添加图例
plt.legend(loc='upper right',fontsize=7)
# 添加网格线
# plt.grid(True)  # x,y 轴都显示
plt.grid(axis='y',alpha=0.1,color='green',linestyle='--')# alpha透明度 #linestyle='--'虚线
# 设置刻度字体大小
plt.xticks(rotation=0,fontsize=9)# rotation旋转角度
plt.yticks(rotation=0,fontsize=9)
# 设置y轴的范围
plt.ylim(0,100)
# 在数据点是显示数据
for x,y in zip(subjects,scores):
    plt.text(x,y,str(y),ha='center',va='bottom',fontsize=10)
# 自动优化排版
plt.tight_layout()

# 显示图表
plt.show()
```

```python
# 条形图
plt.rcParams['font.sans-serif'] = ['SimHei']
# 创建图表, 设置大小
plt.figure(figsize=(7,3))

# 绘图的数据
Countries = ['China','America','Germany','India','Japan']
GDP = [92,85,59,34,4]

# 绘制条形图
plt.barh(Countries,GDP,color='orange')

# 添加标题
plt.title('GDP排名',color='green',fontsize=10)
# 添加坐标轴的标签
plt.xlabel('GDP',color='green',fontsize=10)
plt.ylabel('国家',color='green',fontsize=10)
# 自动优化排版
plt.tight_layout()

# 显示图表
plt.show()
```

```python
# 直方图
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
# 生成三组随机数据
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(2, 1, 1000)
data3 = np.random.normal(-2, 1, 1000)

# 绘制直方图
plt.hist(data1, bins=30, alpha=0.5, label='Data 1') # bins 直方图的箱数, 默认为10
plt.hist(data2, bins=30, alpha=0.5, label='Data 2')
plt.hist(data3, bins=30, alpha=0.5, label='Data 3')

# 设置图表属性
plt.title('正态分布')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# 显示图表
plt.show()
```

```python
# 饼图
plt.rcParams['font.sans-serif'] = ['SimHei']
# 创建图表, 设置大小
plt.figure(figsize=(9,5))

# # 绘图的数据
things = ['学习','娱乐','运动','睡觉','其他']
times = [6,4,2,8,4]
colors = ['red','green','blue','yellow','purple']
# 绘制饼图
plt.pie(times,labels=things,
        autopct='%.1f%%',   # 显示百分比
        startangle=90, # 调整初始画图的角度
        colors=colors
        )
# 添加标题
plt.title('24小时时间分布',color='green',fontsize=10)
# 自动优化排版
plt.tight_layout()

# 显示图表
plt.show()
```

```python
# 环形图
plt.rcParams['font.sans-serif'] = ['SimHei']
# 创建图表, 设置大小
plt.figure(figsize=(9,5))

# # 绘图的数据
things = ['学习','娱乐','运动','睡觉','其他']
times = [6,4,2,8,4]
colors = ['red','green','blue','yellow','purple']
# 绘制环形图
plt.pie(times,labels=things,
        autopct='%.1f%%',   # 显示百分比
        startangle=90, # 调整初始画图的角度
        colors=colors,
        wedgeprops={'width':0.6}, # 挖掉中间占比
        pctdistance=0.7 # 设置百分比数字的位置
        )
# 添加标题
plt.title('24小时时间分布',color='green',fontsize=10)
plt.text(0,0,'总计',ha='center',va='bottom',fontsize=10)
# 自动优化排版
plt.tight_layout()

# 显示图表
plt.show()
```

```python
# 爆炸式饼图
plt.rcParams['font.sans-serif'] = ['SimHei']
# 创建图表, 设置大小
plt.figure(figsize=(5,4))

# 绘图的数据
things = ['学习','娱乐','运动','睡觉','其他']
times = [6,4,2,8,4]
colors = ['red','green','blue','yellow','purple']
explode = [0.1,0,0,0,0] # 设置突出块的位置
# 绘制爆炸式饼图
plt.pie(times,labels=things,
        autopct='%.1f%%',   # 显示百分比
        startangle=0, # 调整初始画图的角度
        colors=colors,
        explode=explode,
        shadow=True,    # 增加阴影
        radius=0.8
        )
# 添加标题
plt.title('24小时时间分布',color='green',fontsize=10)
# 自动优化排版
plt.tight_layout()

# 显示图表
plt.show()
```

```python
# 箱线图
# 箱线图用于展示数据的分布情况, 包括中位数,四分位数以及异常值.
data = {
    '语文':[82,85,88,70,90,76,84,83,95],
    '数学':[75,80,79,93,88,82,87,89,92],
    '英语':[70,72,68,65,78,80,85,90,95]
}
plt.figure(figsize=(8,6))
plt.boxplot(data.values(),tick_labels=data.keys())

plt.title('各科成绩分布')
plt.ylabel('分数')
plt.grid(True,axis='y',linestyle='--',alpha=0.5)
plt.show()
```
